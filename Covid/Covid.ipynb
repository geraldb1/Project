```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid_death = pd.read_excel('CovidDeath.xlsx')
covid_vac = pd.read_excel('CovidVaccination.xlsx')
```

```python
cd_lat = covid_death[covid_death['continent'] == 'South America']
```

```python
cd_lat.describe().T
```

Revisar los datos nulos

```python
import missingno as msg
```

```python
msg.bar(cd_lat,color='blue')
```

Hay datos muy escasos en algunos de las columnas del dato por lo cual se debe eliminar, ya que no genera ningun valor al analisis 

```python
cd_lat.drop(columns = ['icu_patients',
       'icu_patients_per_million', 'hosp_patients',
       'hosp_patients_per_million', 'weekly_icu_admissions',
       'weekly_icu_admissions_per_million', 'weekly_hosp_admissions',
       'weekly_hosp_admissions_per_million'], inplace = True)
```

```python
cd_lat.describe().T
```

Resumen descriptivo tambien de los datos `object`

```python
cd_lat.describe(include='object').T
```

La mayoria de los datos de covid en latinoamerica estan registrados en Argentina

Lo mismo para el tiempo

```python
cd_lat['date'].describe(datetime_is_numeric=True).T
```

La mayoria de los casos registrados ocurren en marzo del 2021, el registro esta dado entre los aÃ±os 2020 y 2022

```python
cd_lat['location'].value_counts(sort=True)
```

Se rellena los datos nulos, como todos los nulos estan en columnas numericas lo rellenare con 0

```python
cd_lat.fillna(0,inplace=True)
```

algunos datos como: continente,codigo que no necesitamos y tambien se va a eliminar 

```python
cd_lat.drop(columns=['iso_code','continent'],inplace=True)
```

```python
plt.figure(figsize=(12,7))
sns.set (rc = {'axes.facecolor':'lightgray'})
sns.countplot(data = cd_lat, x ='location', order= cd_lat['location'].value_counts().index, palette='deep')
plt.xticks(rotation = 60)
plt.title('Registro del Covid contraida por paises',fontsize =20,fontweight = 'heavy',fontfamily = 'Times New Roman')
plt.xlabel('Paises',fontsize = 12, fontweight = 'bold')
plt.ylabel('')
for index, value in enumerate(cd_lat['location'].value_counts().values):
    plt.text(index,value,str(value), ha = 'center', va= 'bottom')
```

Se puede apreciar que el pais que mas lleva registro de los enfermos por covid es Argentina, los otros paises de la region sus registros estan con buen nivel de registro

```python
palette = sns.color_palette("deep", 13)
plt.figure(figsize=(25,10))
sns.set_style('white')
sns.lineplot(data = cd_lat, x = 'date',y = 'new_cases',hue = 'location',style = 'location',palette = palette)
plt.title('Evolucion de los casos registrado por los Paises',fontsize = 15,fontweight = 'bold')
plt.legend(loc = 'upper left')
```

Con mucha diferencia brazil es el pais donde se registra mas nuevos casos de covid en toda latinoamerica

```python
palette = sns.color_palette("deep", 13)
plt.figure(figsize=(25,10))
sns.set_style('white')
sns.lineplot(data = cd_lat, x = 'date',y = 'new_deaths',hue = 'location',style = 'location',palette = palette)
plt.title('Evolucion del registro de nuevos muertos por covid',fontsize = 15,fontweight = 'bold')
plt.legend(loc = 'upper left')
```

Como se podia suponer, brazil al ser el pais que mayor cantidad de enfermos por covid registra tambien registra mayor cantidad de muertes, tamnbien se ve algunos picos que vienen dados por Bolivia y Argentina.

Tambien se observa que en el verano del 2021, Peru tuve un repunte muy alto en sus casos de muertes, comienza a reducir ya conforme acababa el aÃ±o

```python
cd_lat.sample(3)
```

Ahora para poder comparar entre paises e debe tomar la medida de nuevo casos por millon, ya que la cantidad de personas para cada uno es muy diferente

```python
palette = sns.color_palette("deep", 13)
plt.figure(figsize=(25,10))
sns.set_style('white')
sns.lineplot(data = cd_lat, x = 'date',y = 'new_cases_per_million',hue = 'location',style = 'location',palette = palette)
plt.title('Evolucion de los casos registrado por los Paises por millon de habitante',fontsize = 15,fontweight = 'bold')
plt.legend(loc = 'upper left')
```

```python
palette = sns.color_palette("deep", 13)
plt.figure(figsize=(25,10))
sns.set_style('white')
sns.lineplot(data = cd_lat, x = 'date',y = 'new_deaths_per_million',hue = 'location',style = 'location',palette = palette)
plt.title('Evolucion del registro de nuevos muertos por covid dado por millon de Habitante',fontsize = 15,fontweight = 'bold')
plt.legend(loc = 'upper left')
```
